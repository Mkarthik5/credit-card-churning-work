```python
df.drop(columns=["CLIENTNUM","Unnamed: 0"],inplace=True)
```

```python
y=df["Attrition_Flag"]
x=df.drop(columns=["Attrition_Flag"])
```

```python
x=pd.get_dummies(x,columns=["Gender","Education_Level","Income_Category","Card_Category"],drop_first=True)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,stratify=y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_score
s=cross_val_score(DecisionTreeClassifier(),x,y)
print(s)
```

```python
s=cross_val_score(RandomForestClassifier(),x,y)
print(s)
```

```python
s=RandomForestClassifier()
s.fit(x_train,y_train)
```

```python
s.score(x_test,y_test)
```

```python
s.predict(x_test)
```
